```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Question 1
Visualize the evolution of APR distributions over time. Comment on the major finding(s).



**Answer:** The median is approaching 1000 in recent years.APR is getting better and better as time goes by.

```python
url = "https://ncaaorg.s3.amazonaws.com/research/academics/2020RES_APR2019PubDataShare.csv"
df = pd.read_csv(url)
df.head()
```

```python
#turn data into long-form
df_long = pd.melt(df, value_vars=["APR_RATE_2019_1000","APR_RATE_2018_1000","APR_RATE_2017_1000","APR_RATE_2016_1000","APR_RATE_2015_1000","APR_RATE_2014_1000","APR_RATE_2013_1000","APR_RATE_2012_1000","APR_RATE_2011_1000","APR_RATE_2010_1000","APR_RATE_2009_1000","APR_RATE_2008_1000","APR_RATE_2007_1000","APR_RATE_2006_1000","APR_RATE_2005_1000","APR_RATE_2004_1000"],value_name="apr_value")

df_long.head()
```

```python
plt.figure(figsize=(12,6))
apr = sns.boxplot(data=df_long, x= "variable", y="apr_value",medianprops={"color": "coral"}).set(title="APR by year")
# increase the interval of x and y axis
plt.gca().set_yticks(range(350, 1001, 100))
plt.gca().set_xticks(range(0, 16))
plt.gca().set_xticklabels(["2019","2018","2017","2016","2015","2014","2013","2012","2011","2010","2009","2008","2007","2006","2005","2004",],rotation=45)
apr
```
